# Gini

```python
gini = pd.read_csv('/content/drive/MyDrive/economic-inequality-gini-index.csv') 
gini.head()
```

```python
gini =gini.rename(columns={'Entity': 'Country' }) # namen corrigeren. 
```

```python
gini1 = gini[['Country', 'Year','Gini coefficient']]
gini1.head()
```

# GDP

```python
gdp = pd.read_csv('//content/drive/MyDrive/gdp-per-capita-maddison-2020.csv') 
gdp.head()
```

```python
gdp.columns
```

```python
gdp =gdp.rename(columns={'Entity': 'Country', 'GDP per capita': 'GDP'}) # namen corrigeren. 
```

```python
gdp1=gdp[['Country', 'Year', 'GDP']]
gdp1.head()
```

# Beide df samenvoegen

```python
gini_gdp = pd.merge(gini1, gdp1, how="outer", on=['Country','Year'])
gini_gdp.head()
```

```python
gini_gdp1 = gini_gdp.dropna() # evt. aanwezige NaN eruit halen. 
gini_gdp1.head()
```

```python
gini_gdp1.columns
```

```python
correlation =gini_gdp1[['Gini coefficient', 'GDP']]
correlation.head()
```

```python
correlation1 =correlation.corr()
correlation1
```

Conclusie: er is een negatieve correlatie tussen de Gini coefficient en GDP, nl. -0,43 dit betekent dat er een relatief sterke, omgekeerde afhankelijkheid is tussen deze twee waardes.
